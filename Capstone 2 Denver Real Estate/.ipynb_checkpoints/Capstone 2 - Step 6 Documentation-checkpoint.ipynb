```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from pprint import pprint
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, explained_variance_score,mean_absolute_error
import seaborn as sns
import matplotlib.pyplot as plt
import tkinter as tk 
from scipy.stats import gaussian_kde
import pickle
```

```python
import pickle
```

### Load the Data

```python
# load the data saved from step 3
df=pd.read_csv('data\step3_output.csv')
```

### Preprocess data

```python
# only select PROP_CLASS = 1112 (Single Family Residential)
df = df[df.PROP_CLASS == 1112]
# instead of looking at the year the house was built (CCYRBLT), transform it to years since built
df = df.assign(AGE = df.SALE_YEAR-df.CCYRBLT)
# instead of looking at year of remodel (CCAGE_RM), transform it to years since remodel
df=df.assign(RM_AGE = df.SALE_YEAR-df.CCAGE_RM)
# if remodel year is after sale (RM_AGE will be negative), then reset RM_AGE to AGE
df.loc[df.RM_AGE < 0, 'RM_AGE'] = df.AGE
# set remodel age to home age if no remodel year is available
df.loc[df.CCAGE_RM == 0, 'RM_AGE'] = df.AGE
```

```python
# drop columns that are not going to be used
df.drop(['RECEPTION_DATE', 'PIN', 'RECEPTION_NUM', 'SCHEDNUM_x', 'CO_OWNER', 'NBHD_1_CN_y', 'NBHD_1_x'
         , 'SITE_DIR', 'SITE_MODE', 'SITE_MORE', 'SITE_NAME', 'SITE_NBR', 'TAX_DIST', 'ZONE10', 'CD', 'CLASS'
         , 'D_CLASS', 'D_CLASS_CN_x', 'D_CLASS_CN_y', 'GRANTEE', 'GRANTOR', 'INSTRUMENT', 'LEGL_DESCRIPTION'
         , 'PROP_CLASS', 'PROPERTY_CLASS', 'STYLE_CN', 'OWNER', 'OWNER_APT', 'OWNER_CITY', 'OWNER_DIR', 'OWNER_NUM'
         , 'OWNER_ST', 'OWNER_STATE', 'OWNER_TYPE', 'OWNER_ZIP', 'OFCARD', 'ASDLAND', 'ASMT_EXEMPT_AMT'
         , 'ASMT_TAXABLE', 'ASSESS_VALUE', 'TOTAL_VALUE', 'CCYRBLT', 'CCAGE_RM','SALE_PRICE','NBHD_1_CN_x'], axis=1, inplace=True)
# drop any rows with NaN values
df.dropna(inplace=True)
```

```python
# Create feature and target arrays
y = df['IMPROVE_VALUE'].ravel()
X = df[['MKT_CLUS', 
        'LAND_SQFT', 
        'AREA_ABG', 
        'BSMT_AREA', 
        'FBSMT_SQFT', 
        'FULL_B', 
        'HLF_B', 
        'ASMT_APPR_LAND', 
        'AGE', 
        'RM_AGE']]
```

```python
# now scale the X data
scaler = StandardScaler().fit(X)
X_scaled=scaler.transform(X)
```

```python
# Split into training and test set
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state=42)
```

### Set up Random Forest Model 

```python
rf = RandomForestRegressor(n_estimators=100, 
                           max_features=4,
                           bootstrap=False, 
                           max_depth=100, 
                           min_samples_leaf=1, 
                           min_samples_split=2 )
```

```python
model = rf.fit(X_train,y_train)
```

```python
y_pred =model.predict(X_test)
```

```python
# show model results for test data
print('explained variance score: ',explained_variance_score(y_test, y_pred))
print('MAE: ',mean_absolute_error(y_test, y_pred))
```

```python
# reverse transform the test data
X_samples_scaled = pd.DataFrame(X_test[0:2,:],columns = X.columns)
X_samples = scaler.inverse_transform(X_samples_scaled)
```

## Review the model results

```python
sns.regplot(y_test,y_pred)
plt.xlabel('actual')
plt.ylabel('predicted')
plt.title('Figure 02 - Actual vs Predicted Values')
plt.savefig('figures/fig02.png', bbox_inches='tight')
plt.show()
```

```python
feature_importances = pd.DataFrame(model.feature_importances_,
                                   index = X.columns,
                                    columns=['importance']).sort_values('importance',ascending=False)
print(feature_importances)
```

```python
plt.title('Figure 03 - Feature Importances')
plt.barh(feature_importances.index, feature_importances['importance'], color='b', align='center')
plt.xlabel('Relative Importance')
plt.savefig('figures/fig03.png', bbox_inches='tight')
plt.show()
```

```python
# reverse transform the scaled input data
X_test_rev = scaler.inverse_transform(X_test)
```

```python
plt.scatter(X_test_rev[:,2],y_pred)
plt.xlabel('Finished Area Above Ground')
plt.ylabel('Predicted Value')
plt.title('Figure 04 - Predicted Value vs Finished Area Above Ground')
plt.savefig('figures/fig04.png', bbox_inches='tight')
plt.show()
```

```python
# Calculate the point density
x1 = X_test_rev[:,2]
y1 = y_pred
 

xy = np.vstack([x1,y_pred])
z1 = gaussian_kde(xy)(xy)
# Sort the points by density, so that the densest points are plotted last
idx = z1.argsort()
x1 = x1[idx]
y1 = y1[idx]
z1 = z1[idx]

fig, ax = plt.subplots()
ax_=ax.scatter(x1, y1, c=z1, s=50, edgecolor='')
plt.xlabel('Finished Area Above Ground')
plt.ylabel('Predicted Value')
plt.title('Figure 05 - Predicted Value vs Finished Area Above Ground')
plt.colorbar(ax_)
plt.savefig('figures/fig05.png', bbox_inches='tight')
plt.show()
```

```python
# make a dataframe with test data and predicted value
dfrev = pd.DataFrame(X_test_rev, columns=X.columns)
dfrev['y_pred']=y_pred
```

```python
# plot distribution of predicted value versus number of full bathrooms                   
fontsize = 10

fig, axes = plt.subplots()

sns.violinplot('FULL_B','y_pred', data=dfrev, ax = axes)
axes.set_title('Figure 06 - Predicted Value vs Full Bathrooms')

axes.yaxis.grid(True)
axes.set_xlabel('Full Bathrooms')
axes.set_ylabel('Predicted Value')

plt.savefig('figures/fig06.png', bbox_inches='tight')
plt.show()
```

```python
# tkinter GUI
root= tk.Tk()

canvas01 = tk.Canvas(root, width = 500, height = 350)
canvas01.pack()

# 01 - Market Cluster
label01 = tk.Label(root, text='Market Cluster:')
canvas01.create_window(200, 40, window=label01)

entry01 = tk.Entry (root)
canvas01.create_window(370, 40, window=entry01)

# 02 - Land Square Feet 
label02 = tk.Label(root, text='Land (Square Feet):')
canvas01.create_window(200, 60, window=label02)

entry02 = tk.Entry (root)
canvas01.create_window(370, 60, window=entry02)

# 03 - Above Ground Square Feet
label03 = tk.Label(root, text='Finished Above Ground (Square Feet):')
canvas01.create_window(200, 80, window=label03)

entry03 = tk.Entry (root)
canvas01.create_window(370, 80, window=entry03)

# 04 - Basement Square Feet
label04 = tk.Label(root, text='Basement (Square Feet)')
canvas01.create_window(200, 100, window=label04)

entry04 = tk.Entry (root)
canvas01.create_window(370, 100, window=entry04)

# 05 - Finished Basement Square Feet 
label05 = tk.Label(root, text='Finished Basement (Square Feet):')
canvas01.create_window(200, 120, window=label05)

entry05 = tk.Entry (root)
canvas01.create_window(370, 120, window=entry05)

# 06 - Number of Full Bathrooms 
label06 = tk.Label(root, text='Number of Full Bathrooms:')
canvas01.create_window(200, 140, window=label06)

entry06 = tk.Entry (root)
canvas01.create_window(370, 140, window=entry06)

# 07 - Number of Half Bathrooms
label07 = tk.Label(root, text='Number of Half Bathrooms:')
canvas01.create_window(200, 160, window=label07)

entry07 = tk.Entry (root)
canvas01.create_window(370, 160, window=entry07)

# 08 - Assessed Value of Land
label08 = tk.Label(root, text='Assessed Value of Land:')
canvas01.create_window(200, 180, window=label08)

entry08 = tk.Entry (root)
canvas01.create_window(370, 180, window=entry08)

# 09 - Age of home
label09 = tk.Label(root, text='Age of home:')
canvas01.create_window(200, 200, window=label09)

entry09 = tk.Entry (root)
canvas01.create_window(370, 200, window=entry09)

# 10 - Age of home since last remodel
label10 = tk.Label(root, text='Age of home since last remodel:')
canvas01.create_window(200, 220, window=label10)

entry10 = tk.Entry (root)
canvas01.create_window(370, 220, window=entry10)


def values(): 
    global MKT_CLUS 
    MKT_CLUS = float(entry01.get())

    global LAND_SQFT 
    LAND_SQFT = float(entry02.get())

    global AREA_ABG 
    AREA_ABG = float(entry03.get())

    global BSMT_AREA 
    BSMT_AREA = float(entry04.get())

    global FBSMT_SQFT 
    FBSMT_SQFT = float(entry05.get())

    global FULL_B 
    FULL_B = float(entry06.get())

    global HLF_B 
    HLF_B = float(entry07.get())

    global ASMT_APPR_LAND 
    ASMT_APPR_LAND = float(entry08.get())

    global AGE 
    AGE = float(entry09.get())

    global RM_AGE 
    RM_AGE = float(entry10.get())
    
    Request_unscaled = [[MKT_CLUS, 
                         LAND_SQFT, 
                         AREA_ABG, 
                         BSMT_AREA, 
                         FBSMT_SQFT, 
                         FULL_B, 
                         HLF_B, 
                         ASMT_APPR_LAND, 
                         AGE, 
                         RM_AGE]]
    
    Request_scaled = scaler.transform(Request_unscaled)

    
    Prediction_result  = ('  Predicted Result: ', model.predict(Request_scaled))
    label_Prediction = tk.Label(root, text= Prediction_result, bg='sky blue')
    canvas01.create_window(370, 320, window=label_Prediction)
    
button1 = tk.Button (root, text='      Predict      ',command=values, bg='green', fg='white', font=11)
canvas01.create_window(370, 270, window=button1)
 
root.mainloop()
```
